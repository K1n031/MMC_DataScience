```python
import numpy as np
np.random.seed(42)
```

# Problem 1: Tạo hai ma trận 3x3 với các phần tử integer ngẫu nhiên từ 1 - 10.

```python
# Your code here
np_array_1 = np.random.randint(1, 10, size=(3, 3))
np_array_2 = np.random.randint(1, 10, size=(3, 3))

print("Ma trận 1:", np_array_1)
print("Ma trận 2:", np_array_2)
```

# Problem 2: Thực hiện phép cộng phần tử, nhân từng phần tử, tích vô hướng với hai ma trận vừa được tạo.


```python
# Your code here
sum_array = np_array_1 + np_array_2
print("Tổng 2 ma trận:", sum_array)

multi_array = np_array_1 * np_array_2
print("Nhân 2 ma trận:", multi_array)

dot_product = np.dot(np_array_1, np_array_2)
print("Tích vô hướng 2 ma trận:", dot_product)
```

# Problem 3: VIết hàm nhận vào một mảng Numpy, sau đó trả về tổng tất cả các phần tử của chúng


```python
# Your code here
def sum_elements(matrix):
    return np.sum(matrix)

matrix = np.random.randint(1, 10, size=(3, 3))
print("Ma trận:", matrix)
print("Tổng tất cả các phàn tử của ma trận:", sum_elements(matrix))
```
