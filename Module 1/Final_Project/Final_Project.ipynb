### Task 2
Nhiệm vụ của bạn cho phần này của chương trình là thực hiện như sau:

2.1. Báo cáo tổng số dòng dữ liệu được lưu trữ trong tệp.

2.2. Báo cáo tổng số dòng dữ liệu không hợp lệ trong tệp.

Một dòng hợp lệ chứa danh sách 26 giá trị được phân tách bằng dấu phẩy N# cho một học sinh là mục đầu tiên trên dòng. Nó phải chứa ký tự “N” theo sau là 8 ký tự số.

2.3. Nếu một dòng dữ liệu không hợp lệ, bạn nên báo cáo cho người dùng bằng cách in ra một thông báo lỗi.

Gợi ý: Sử dụng phương pháp split để tách dữ liệu ra khỏi tệp.

```python
def analyzeClass():
    while True:
        filename = input("Enter name class: ") 
        FILE_PATH = "./Data_Files/" + filename + ".txt"
        try:
            f = open(FILE_PATH,"r+")
        except Exception as e:
            print(e)
        else:
            print("\n SUCCESSFULLY OPENED",filename)
            
        print("**** ANALYZING ****")
        k = 0 # number of lines
        error = 0 # number of lines error
        
        for line in f.readlines(): # Vòng lặp kiểm tra các lỗi của dữ liệu
            k += 1
            listValue = line.rstrip("\n").split(",")
            if len(listValue) != 26:
                print("Invalid line of data: does not contain exactly 26 values: \n")
                print(line,"\n")
                error += 1
            elif not listValue[0][1:].isdigit() or len(listValue[0]) !=9:
                print("Invalid line of data: N# is invalid \n")
                print(line, "\n")
                error += 1
        if error == 0:
            print("No errors found!\n")
        
        print("**** REPORT ****\n")
        print("Total valid lines of data: ", k - error,"\n")
        print("Total invalid lines of data: ", error,"\n")

        choice = input("Do you want nhập thêm không? (0: không, 1: có)")
        
        if choice == "0":
            print("Goodbye!")
            break
```

```python
analyzeClass()
```

### Task 3
Chúng ta cũng cần thống kê các yêu cầu sau:

Đếm số lượng học sinh đạt điểm cao (>80.

Điểm trung bnh.

Điểm cao nhất.

Điểm thấp nhất.

Miền giá trị của điểm (cao nhất trừ tấp nhất).

Giá trị trung vị (Sắp xếp các điểm theo thứ tự tăng dần. Nếu # học sinh là số lẻ, bạn có thể lấy giá trị nằm ở giữa của tất cả các điểm (tức là [0, 50, 100] — trung vị là 50). Nếu # học sinh là chẵn bạn có thể tính giá trị trung vị bằng cách lấy giá trị trung bình của hai giá trị giữa (tức là [0, 50, 60, 100] — giá trị trungvị là 55)).

Trả về các câu hỏi bị học sinh bỏ qua nhiều nhất theo thứ tự: số thứ tự câu hỏi - số lượng học sinh bỏ qua - tỉ lệ bị bỏ qua (nếu có cùng số lượng cho nhiều câu hỏi bị bỏ thì phải liệt ê ra đầy đủ).

Trả về các câu hỏi bị học sinh sai qua nhiều nhất theo thứ tự: số thứ tự câu hỏi - số lượng học sinh trả lời sai - tỉ lệ bị sai (nếu có cùng số lượng cho nhiều câu hỏi bị sai thì phải liệt kê ra đầy đủ).

```python
while True:
    filename = input("Enter name class: ")
    FILE_PATH = "./Data_files/" + filename + ".txt"
    try:
        f = open(FILE_PATH, "r+")
    except Exception as e:
        print(e)
    else:
        print("File opened successfully")
    
    print("**** ANALYZING ****")
    k = 0 # Number of lines
    error = 0 # Number of lines error
    grade = [] # List grade
    valid_lines = [] # Number of valid lines
    answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D".split(",")

    for line in f.readlines():
        k += 1
        listValue = line.rstrip("\n").split(",")
        if len(listValue) != 26:
            print("Invalid line of data: does not contain exactly 27 values: \n")
            print(line, "\n")
            error += 1
        elif not listValue[0][1:].isdigit() or len(listValue[0]) != 9:
            print("Invalid line of data: N# is invalid \n")
            print(line, "\n")
            error += 1
        else:
            valid_lines.append(listValue)

    for line in valid_lines:
        answer = line[1:]
        score = 0
        for i in range(25):
            if answer[i] == answer_key[i]:
                score += 4
            elif answer[i] == "":
                score += 0
            else:
                score -= 1
        grade.append(score)

    if error == 0:
        print("No errors found!\n")

    print("**** REPORT ****\n")
    print("Total valid lines of data: ", k - error, "\n")
    print("Total invalid lines of data: ", error, "\n")
    print("Mean (average) score: ", sum(grade) / len(grade), "\n")
    print("Highest score: ", max(grade), "\n")
    print("Lowest score: ", min(grade), "\n")
    print("Range of scores: ", max(grade) - min(grade), "\n")
    
    s = sorted(grade)
    n = len(grade)

    if n % 2 == 1:
        median = s[n//2]
    else:
        median = sum(s[n//2-1:n//2+1])/2
    print("Median score: ", median, "\n")

    choice = input("Do you want to analyze another class? (0: no, 1: yes)")
    if choice == "0":
        print("Goodbye!")
        break
```

### Task 4
Cuối cùng, yêu cầu chương trình là tạo một tệp “kết quả” chứa các kết quả chi tiết cho từng học sinh trong lớp. Mỗi dòng của tệp này phải chứa số ID của học sinh, dấu phẩy và sau đó là điểm của họ. Bạn nên đặt tên tệp này dựa trên tên tệp gốc được cung cấp — ví dụ: nếu người dùng chọn “class1.txt”, bạn nên lưu trữ kết quả trong tệp có tên “class1_grades.txt”.

```python
while True:
    filename = input("Enter name class: ")
    FILE_PATH = "./Data_files/" + filename + ".txt"
    try:
        f = open(FILE_PATH, "r+")
    except Exception as e:
        print(e)
    else:
        print("File opened successfully")

    print("**** ANALYZING ****")

    student_number = []     # List of student numbers   
    grade = []              # List of grades
    valid_lines = []        # List of valid lines

    answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D".split(",")

    for line in f.readlines():
        listValue = line.rstrip("\n").split(",")

        if len(listValue) == 26 and listValue[0][1:].isdigit() and len(listValue[0]) == 9:
            valid_lines.append(listValue)

    for line in valid_lines:
        student_number.append(line[0])
        answer = line[1:]
        score = 0

        for i in range(25):
            if answer[i] == answer_key[i]:
                score += 4
            elif answer[i] == "":
                score += 0
            else:
                score -= 1
        grade.append(score)

    FILE_PATH_GRADE = "./Data_files/" + filename + "_grade.txt"

    with open(FILE_PATH_GRADE, "w") as f_grade:
        for student_number, grade in zip(student_number, grade):
            f_grade.write(student_number + "," + str(grade) + "\n")
        print("Successfully created grade file for: ", filename)
        f.close()
        choice = input("Do you want to analyze another class? (0: no, 1: yes):")
        if choice == "0":
            print("Goodbye!")
            break
```
